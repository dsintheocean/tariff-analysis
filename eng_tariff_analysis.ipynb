### Prepare the data

**Here are the columns**

- `reg_date` from the `users` table
- `churn_date` from the `users` table
- `call_date` from the `calls` table
- `message_date` from the `messages` table
- `session_date` from the `sessions` table

**to a new type using the `to_datetime()` method.**

````python
users['reg_date'] = pd.to_datetime(users['reg_date'], format='%Y-%m-%d', errors='ignore') # processing the reg_date column

users['churn_date'] = pd.to_datetime(users['churn_date'], format='%Y-%m-%d', errors='ignore') # processing churn_date column

calls['call_date'] = pd.to_datetime(calls['call_date'], format='%Y-%m-%d', errors='ignore')# processing call_date column

messages['message_date'] = pd.to_datetime(messages['message_date'], format='%Y-%m-%d', errors='ignore')# processing message_date column

sessions['session_date'] = pd.to_datetime(sessions['session_date'], format='%Y-%m-%d', errors='ignore') # processing session_date column
````

**The data contains calls with zero duration. This is not an error: zeros indicate missed calls, so they do not need to be deleted.**

**However, in the `duration` column of the `calls` dataframe, the values are fractional. Round the values of the `duration` column up using the `numpy.ceil()` method and cast the `duration` column to the `int` type.**

````python
calls['duration'] = np.ceil(calls['duration']).astype(int) # round duration column values using np.ceil() and cast to int
````

**Remove the `Unnamed: 0` column from the `sessions` dataframe. A column with this name appears when data is saved with an index specified (`df.to_csv(..., index=column)`). We won't need it now.**

````python
sessions = sessions[['id', 'mb_used', 'session_date', 'user_id']]
````

**Create a `month` column in the `calls` dataframe with the month number from the `call_date` column.**

````python
calls['month'] = pd.DatetimeIndex(calls['call_date']).month
````

**Create a `month` column in the `messages` dataframe with the month number from the `message_date` column.**

````python
messages['month'] = pd.DatetimeIndex(messages['message_date']).month
````

**Create a `month` column in the `sessions` dataframe with the month number from the `session_date` column.**

````python
sessions['month'] = pd.DatetimeIndex(sessions['session_date']).month
````

**Let's count the number of conversation calls made for each user by month.**

````python
calls_per_month = calls.groupby(['user_id', 'month']).agg(calls=('duration', 'count'))
````

````python
calls_per_month.head(30) # print the first 30 lines to the screen
````

**Let's calculate the number of spent conversation minutes for each user by month and save it in the `minutes_per_month` variable. For this you will need:**

- group the dataframe with information about calls into two columns - with user IDs and month numbers;
- after grouping, select the `duration` column
- then apply the method to calculate the amount.

**Print the first 30 lines of `minutes_per_month`.**

````python
minutes_per_month = calls.groupby(['user_id', 'month']).agg(minutes=('duration', 'sum')) # counting minutes spent for each user by month
````

````python
minutes_per_month.head(30) # print first 30 lines to screen
````

**Let's count the number of messages sent by month for each user and save it in the `messages_per_month` variable. For this you will need:**

- group the dataframe with information about messages into two columns - with user IDs and month numbers;
- after grouping, select the `message_date` column;
- then apply the method to calculate the quantity.

**Print the first 30 lines of `messages_per_month`.**

````python
messages_per_month = messages.groupby(['user_id', 'month']).agg(messages=('message_date', 'count'))# counting the number of messages sent for each user by month
````

````python
messages_per_month.head(30) # print the first 30 lines to the screen
````

**Let's calculate the number of megabytes spent by month for each user and save it in the `sessions_per_month` variable. For this you will need:**

- group the dataframe with information about messages into two columns - with user IDs and month numbers;
- then apply the method to calculate the sum: `.agg({'mb_used': 'sum'})`

````python
sessions_per_month = sessions.groupby(['user_id', 'month']).agg({'mb_used': 'sum'}) # counting megabytes spent for each user by month
````

````python
sessions_per_month.head(30) # print the first 30 lines to the screen
````

### Data analysis and revenue calculation

We combine all the values calculated above into one dataframe `user_behavior`.
For each user-month pair, information about the tariff, number of calls, messages and megabytes spent will be available.

````python
users['churn_date'].count() / users['churn_date'].shape[0] * 100
````

7.6% of clients from the dataset terminated their contracts

````python
user_behavior = calls_per_month\
    .merge(messages_per_month, left_index=True, right_index=True, how='outer')\
    .merge(sessions_per_month, left_index=True, right_index=True, how='outer')\
    .merge(minutes_per_month, left_index=True, right_index=True, how='outer')\
    .reset_index()\
    .merge(users, how='left', left_on='user_id', right_on='user_id')\

user_behavior.head()
````

Let's check for gaps in the `user behavior` table after the merge:

````python
user_behavior.isna().sum()
````

Let's fill in the resulting gaps in the data:

````python
user_behavior['calls'] = user_behavior['calls'].fillna(0)
user_behavior['minutes'] = user_behavior['minutes'].fillna(0)
user_behavior['messages'] = user_behavior['messages'].fillna(0)
user_behavior['mb_used'] = user_behavior['mb_used'].fillna(0)
````

Add information about tariffs

````python
# renaming the tariff_name column to the simpler tariff
tariffs = tariffs.rename(
     columns={
         'tariff_name': 'tariff'
     }
)
````

````python
user_behavior = user_behavior.merge(tariffs, on='tariff')
````

We count the number of conversation minutes, messages and megabytes exceeding those included in the tariff

````python
user_behavior['paid_minutes'] = user_behavior['minutes'] - user_behavior['minutes_included']
user_behavior['paid_messages'] = user_behavior['messages'] - user_behavior['messages_included']
user_behavior['paid_mb'] = user_behavior['mb_used'] - user_behavior['mb_per_month_included']

for col in ['paid_messages', 'paid_minutes', 'paid_mb']:
    user_behavior.loc[user_behavior[col] < 0, col] = 0
````

We convert megabytes exceeding the tariff into gigabytes and save them in the `paid_gb` column

````python
user_behavior['paid_gb'] = np.ceil(user_behavior['paid_mb'] / 1024).astype(int)
````

We calculate revenue for call minutes, messages and internet

````python
user_behavior['cost_minutes'] = user_behavior['paid_minutes'] * user_behavior['rub_per_minute']
user_behavior['cost_messages'] = user_behavior['paid_messages'] * user_behavior['rub_per_message']
user_behavior['cost_gb'] = user_behavior['paid_gb'] * user_behavior['rub_per_gb']
````

We calculate the monthly revenue from each user, it will be stored in the `total_cost` column

````python
user_behavior['total_cost'] = \
      user_behavior['rub_monthly_fee']\
    + user_behavior['cost_minutes']\
    + user_behavior['cost_messages']\
    + user_behavior['cost_gb']
````

The `stats_df` dataframe for each month-tariff pair will store the main characteristics

````python
# saving statistical metrics for each month-tariff pair
# in one table stats_df (mean, standard deviation, median)

stats_df = user_behavior.pivot_table(
             index=['month', 'tariff'],\
             values=['calls', 'minutes', 'messages', 'mb_used'],\
             aggfunc=['mean', 'std', 'median']\
).round(2).reset_index()

stats_df.columns=['month', 'tariff', 'calls_mean', 'sessions_mean', 'messages_mean', 'minutes_mean',
                                      'calls_std', 'sessions_std', 'messages_std', 'minutes_std',
                                      'calls_median', 'sessions_median', 'messages_median', 'minutes_median']

stats_df.head(10)
````

### Testing hypotheses

**Let’s check the hypotheses: the average revenue of users of the “Ultra” and “Smart” tariffs differ;**

```
H_0: Revenue (total_cost) of "Ultra" users = revenue (total_cost) of "Smart" users`
H_a: Revenue (total_cost) of "Ultra" users ≠ revenue (total_cost) of "Smart" users`
alpha = 0.05
```

````python
results = st.ttest_ind(
     user_behavior[user_behavior['tariff'] =='smart']['total_cost'],
     user_behavior[user_behavior['tariff'] =='ultra']['total_cost'],
     equal_var=False) #method call to test the hypothesis

alpha = 0.05 #set the significance level value

print(results.pvalue) # print the p-value to the screen
if results.pvalue < alpha:
     print('Reject the null hypothesis')
else:
     print('Failed to reject the null hypothesis')
     # conditional statement with response string output
````

**Let's test the hypotheses: users from Moscow bring in more revenue than users from other cities;**

```
H_0: Revenue (total_cost) of users from Moscow = revenue (total_cost) of users not from Moscow`
H_1: Revenue (total_cost) of users from Moscow ≠ revenue (total_cost) of users not from Moscow`
alpha = 0.05
```

````python
results = st.ttest_ind(
     user_behavior[user_behavior['city'] == 'Москва']['total_cost'],
     user_behavior[user_behavior['city'] != 'Москва']['total_cost'],
     equal_var=False) #method call to test the hypothesis

alpha = 0.05 #set the significance level value

print(results.pvalue) # print the p-value to the screen
if results.pvalue < alpha:
     print('Reject the null hypothesis')
else:
     print('Failed to reject the null hypothesis')
     # conditional statement with response string output
````
